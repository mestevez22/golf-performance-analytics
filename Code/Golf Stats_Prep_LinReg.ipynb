```python
#function to add shot type as per KR's request 
def add_shot(x):
    if x['Launch Direction'] > -1.5 and x['Launch Direction'] < 1.5: 
        return 'Straight'
    elif x['Launch Direction'] < -1.5: 
        return 'Left'
    else: 
        return 'Right'

#test = pd.read_csv('Desktop/Kavis Golf Stats/mlm2pro_shotexport_022424.csv', header =1)
#apr10['Shot Type'] = apr10.apply(add_shot)
```

```python
#function to clean up df
def clean_pd(df):
    clean = df.copy()
    clean = clean.dropna()
    clean = clean.drop(clean[clean.iloc[:, 0] == 'Club Type'].index).reset_index(drop = True)
    clean['Launch Direction'] = clean['Launch Direction'].astype(float)
    clean['Shot Type'] = clean.apply(add_shot, axis =1)
    return clean 

#demo = pd.read_csv('Golf 721/mlm2pro_shotexport_022724 copy.csv',header = 1)
#demo
```

```python
df_dummies = df_dummies.drop(columns = ['Spin Rate', 'Spin Axis'])
X = df_dummies.iloc[:, 4:]
X = X.astype('float')
y = df_dummies['Carry Distance']
X
```

```python
from sklearn import linear_model, model_selection, preprocessing, datasets
scaler = preprocessing.StandardScaler()
```

```python
X_train, X_test,\
y_train, y_test = train_test_split(X, y,
                                   test_size=0.3,
                                   random_state=10)

scaler.fit(X_train)

#Transforming all data
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
lm.score(X_test,y_test)
#print('Coefficients: ', reg.coef_)
#coef = pd.DataFrame(model.coef_, X.columns, model.tvalues, columns = ['Coefficient','P Value'])
#coef
```
